#EJERCICIO 1
Valor esperado de Annie en el torneo de póquer


```python
# Datos del problema
costo_inscripcion = 10000  # en dólares
premio = 500000  # en dólares
probabilidad_ganar = 0.005  # 0.5%

# Cálculo del valor esperado
valor_esperado = (probabilidad_ganar * premio) - costo_inscripcion

print(f"El valor esperado de que Annie se inscriba en el torneo es: ${valor_esperado:.2f}")
```

#EJERCICIO 2 Costo total esperado del viaje en tren de Alice



```python
# Datos del problema
costo_boleto = 5  # en euros
multa = 300  # en euros
probabilidad_revision = 0.05  # 5%

# Cálculo del costo total esperado
costo_esperado = costo_boleto + (probabilidad_revision * multa)

print(f"El costo total esperado del viaje en tren de Alice es: €{costo_esperado:.2f}")
```

#EJERCICIO 3 Valor esperado de un dado justo de 6 caras



```python
# Datos del problema
resultados_dado = [1, 2, 3, 4, 5, 6]
probabilidad = 1 / len(resultados_dado)

# Cálculo del valor esperado
valor_esperado = sum(resultado * probabilidad for resultado in resultados_dado)

print(f"El valor esperado de un dado justo de 6 caras es: {valor_esperado}")
```

EJERCICIO 4 Rendimiento esperado de la apuesta de Ivy

```python
# Datos del problema
ganancia = 10  # en dólares
perdida = 11  # en dólares
probabilidad_ganar = 0.5  # 50%

# Cálculo del rendimiento esperado
rendimiento_esperado = (probabilidad_ganar * ganancia) - ((1 - probabilidad_ganar) * perdida)

print(f"El rendimiento esperado de hacer esta apuesta es: ${rendimiento_esperado:.2f}")
```

#Unidad 3. Variables aleatorias

#Ejercicio 1: Gráfica de frecuencia relativa del clima

```python
import matplotlib.pyplot as plt
from collections import Counter

# Datos del clima
clima_dias = [
    "SSS", "SSS", "SNN", "SNS", "NNS",
    "SSS", "SSN", "NNS", "SSS", "NSS"
]

# Contar el número de momentos soleados por día
soleados_por_dia = [clima.count('S') for clima in clima_dias]

# Calcular la frecuencia relativa
frecuencia = Counter(soleados_por_dia)
total_dias = len(clima_dias)
frecuencia_relativa = {k: v / total_dias for k, v in frecuencia.items()}

# Ordenar los resultados
frecuencia_relativa_ordenada = dict(sorted(frecuencia_relativa.items()))

# Graficar la frecuencia relativa
plt.bar(frecuencia_relativa_ordenada.keys(), frecuencia_relativa_ordenada.values(), color='skyblue')
plt.xlabel('Número de momentos soleados en un día')
plt.ylabel('Frecuencia relativa')
plt.title('Frecuencia relativa de momentos soleados en un día')
plt.xticks(list(frecuencia_relativa_ordenada.keys()))
plt.show()
```

#Ejercicio 2: Gráfica de frecuencia relativa de plantas sobrevivientes

```python
import matplotlib.pyplot as plt
from collections import Counter

# Datos de las plantas
lugares = [
    "DSD", "DSS", "SSS", "DSD", "SDS"
]

# Contar el número de plantas sobrevivientes por ubicación
sobrevivientes_por_lugar = [lugar.count('S') for lugar in lugares]

# Calcular la frecuencia relativa
frecuencia = Counter(sobrevivientes_por_lugar)
total_lugares = len(lugares)
frecuencia_relativa = {k: v / total_lugares for k, v in frecuencia.items()}

# Ordenar los resultados
frecuencia_relativa_ordenada = dict(sorted(frecuencia_relativa.items()))

# Graficar la frecuencia relativa
plt.bar(frecuencia_relativa_ordenada.keys(), frecuencia_relativa_ordenada.values(), color='lightgreen')
plt.xlabel('Número de plantas sobrevivientes por ubicación')
plt.ylabel('Frecuencia relativa')
plt.title('Frecuencia relativa de plantas sobrevivientes por ubicación')
plt.xticks(list(frecuencia_relativa_ordenada.keys()))
plt.show()
```
