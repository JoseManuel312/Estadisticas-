```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot

#Cargar el archivo CSV en una variable llamada covid
covid = pd.read_csv("/content/20-06-30-COVID19MEXICO - codificado.zip", encoding="ISO-8859-1")

#Mostrar los primeros 5 registros del archivo
covid.head()
```

```python
# Mostramos los primeros 5 registro del archivo
covid.head()
```

```python
# Mostrar la informacion del archivo.
# Muestra el numero de registros, tipos de datos de la columna, valores nulos, memoria util
covid.info()
```

```python
# Mostrar la edad maxima de la base de datos de la columna edad
covid.EDAD.max()
```

```python
# Mostrar la edad minima de la base de datos de la columna edad
covid.EDAD.min()
```

```python
# Mostrar los estadisticos descriptivos de la columna edad
covid.EDAD.describe()
```

```python
# Dibuja el histograma del atributo "Tipo de paciente", esto permite
# comparar cuantos hay de cada grupo
plt.pyplot.hist(covid.TIPO_PACIENTE, bins = 3)


# Establece las etiquetas x/y y muestra el titulo
plt.pyplot.title("Histograma de tipos Pacientes en Prueba COVID")
plt.pyplot.xlabel("Tipo de Paciente")
plt.pyplot.ylabel("Frecuencia")
```

```python
# Crea un diagrama de cajas para visualizar la distribucion
# de los rangos de edad por sexo
sns.boxplot(x="SEXO", y="EDAD", data=covid)
```

```python
# Crea un diagrama de cajas para visualizar la distribucion
# de los rangos de edad por tipo de paciente
sns.boxplot(x="TIPO_PACIENTE", y="EDAD", data=covid)
```
